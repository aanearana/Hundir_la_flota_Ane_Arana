```python
#Crea la funci√≥n `crear_tablero(tama√±o)`, un tablero por defecto de 10x10 relleno del car√°cter "_" con numpy.
import numpy as np

def crear_tablero(tama√±o, caracter):
    #Tablero Player 01
    tablero = np.full((tama√±o, tama√±o), caracter)
    return tablero

#El barco debera de llevar una fila y una columna. Para la extension de ella debera de llevar tuplas. (0,0)
def colocar_barco(barco, tablero):
    #Barco como elemento para posicionar, tablero elemento donde colocar
    #Los argumentos de la anchura deveran de llevar las coordenada x.
    #Los argumentos de la altura devera de llevar la coordenada y.
    for fila, columna in barco:
        tablero[fila, columna] = "O"
    return tablero


def disparar(casilla, tablero):
    fila, columna = casilla
    if tablero[fila, columna] == "O":
        tablero[fila, columna] = "‚ò†"
        print("¬°Tocado!")
    elif tablero[fila, columna] == "_":
        tablero[fila, columna] = "‚Ä¢"
        print("Agua")
    if not (0 <= fila < tablero.shape[0] and 0 <= columna < tablero.shape[1]):
        print(f"Error: Coordenadas {casilla} est√°n fuera del tablero del tablero.")
    else:
        print("Ya has disparado aqu√≠")
    return tablero


#Player 01 - Tablero 
tablero_pl_1 = crear_tablero(10, "_")
#\n para salto de l√≠nea
print("El Tablero Player 01\n", tablero_pl_1)

#Player 02 - Tablero
tablero_pl_2 = crear_tablero(10, "_")
print("El Tablero Player 02\n",  tablero_pl_2)

```
